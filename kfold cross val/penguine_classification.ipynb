Palmer Archipelago (Antarctica) penguin data


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import tree 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
df=pd.read_csv('/content/penguins_lter.csv')
df.head()
```

```python
data=df[['Species','Island','Culmen Length (mm)','Culmen Depth (mm)','Flipper Length (mm)','Body Mass (g)','Sex']]
data.head()
```

```python
data.shape
```

```python
data['Species'].value_counts()
```

```python
data['Island'].unique()
```

```python
data.isnull().sum()
```

```python

data.drop(data[data['Body Mass (g)'].isnull()].index,axis=0, inplace=True)
```

```python
data['Sex'] = data['Sex'].fillna('MALE')
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
dummies=pd.get_dummies(data['Sex'])
```

```python
df=pd.concat([dummies,data],axis='columns')
df.head()
```

```python
final=df.drop(['Sex','FEMALE'],axis='columns')
final.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = final
dfle.Species = le.fit_transform(dfle.Species)
dfle
```

```python
dfle1 = dfle
dfle1.Island= le.fit_transform(dfle.Island)
dfle1
```

```python
x=dfle1[['MALE','Island','Culmen Length (mm)','Culmen Depth (mm)','Flipper Length (mm)','Body Mass (g)']]
```

```python
y=dfle1['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score1=cross_val_score(LogisticRegression(),x,y,cv=3)
np.average(score1)
```

```python
score2=cross_val_score(tree.DecisionTreeClassifier(),x,y,cv=3)
np.average(score2)
```

```python
score3=cross_val_score(RandomForestClassifier(),x,y,cv=3)
np.average(score3)
```

```python
score4=cross_val_score(SVC(),x,y,cv=3)
np.average(score4)
```
